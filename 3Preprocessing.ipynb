```python
import pandas as pd
# Set the maximum number of rows to display
pd.set_option('display.max_rows', None)
```

```python
# Load the downloaded data
nifty_data = pd.read_csv("feature_extraction_nifty_data.csv")
nifty_data.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close_Prev_Day', 'Gap', 'Day_Num',	'SMA_5', 'SMA_10', 'UpDown']
nifty_data.set_index('Date', inplace=True)
```

```python
X = nifty_data.iloc[:, :-1]  # All rows, all columns except the last
y = pd.Series(nifty_data.iloc[:, -1])   # All rows, the last column
```

```python
y
```

```python
# Determine split indices
total_days = len(nifty_data)
train_size = int(0.7 * total_days)
val_size = int(0.15 * total_days)

# Split the data
X_train = X.iloc[:train_size]
X_val = X.iloc[train_size:train_size + val_size]
X_test = X.iloc[train_size + val_size:]

y_train = y.iloc[:train_size]
y_val = y.iloc[train_size:train_size + val_size]
y_test = y.iloc[train_size + val_size:]

# Print the sizes of each split
print(f"Training Data: {X_train.shape[0]} days and training target:{y_train.shape[0]}")
print(f"Validation Data: {X_val.shape[0]} days and validation target:{y_val.shape[0]}")
print(f"Test Data: {X_test.shape[0]} days and test target:{y_test.shape[0]}")
```

```python
#encoding

#there is no need to encode categorical data in our case, since 'upDown' is target and 'Day_Num' is already label encoded

numerical_features = ['Open', 'High', 'Low', 'Close', 'Volume', 'Close_Prev_Day', 'Gap', 'SMA_5', 'SMA_10']
ordinal_features = ['Day_Num']
nominal_features = []

# Separate numerical and categorical data
X_train_numerical = X_train[numerical_features]
X_train_ordinal = X_train[ordinal_features]
X_train_nominal = X_train[nominal_features]

X_val_numerical = X_val[numerical_features]
X_val_ordinal = X_val[ordinal_features]
X_val_nominal = X_val[nominal_features]

X_test_numerical = X_test[numerical_features]
X_test_ordinal = X_test[ordinal_features]
X_test_nominal = X_test[nominal_features]
```

```python
# Scale numerical features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_numerical = scaler.fit_transform(X_train_numerical)

# Transform the validation and test data using the fitted scaler
X_val_numerical = scaler.transform(X_val_numerical)
X_test_numerical = scaler.transform(X_test_numerical)
```

```python
# Combine scaled numerical and encoded categorical features
import numpy as np
X_train = np.hstack((X_train_numerical, X_train_ordinal, X_train_nominal))
X_val = np.hstack((X_val_numerical, X_val_ordinal, X_val_nominal))
X_test = np.hstack((X_test_numerical, X_test_ordinal, X_test_nominal))
```

```python
# change processed data back to dataframe from a numparray:

# Column names for numerical features (already scaled)
numerical_column_names = numerical_features  # These are the names you used for numerical columns

# Column name for the ordinal feature (Day_Num)
ordinal_column_names = ordinal_features

# Column name for the nominal feature (UpDown)
nominal_column_names = nominal_features

# Combine all feature names
all_column_names = numerical_column_names + ordinal_column_names + nominal_column_names

# Convert the processed numpy arrays back into DataFrames
X_train = pd.DataFrame(X_train, columns=all_column_names)
X_val = pd.DataFrame(X_val, columns=all_column_names)
X_test = pd.DataFrame(X_test, columns=all_column_names)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier

# Initialize and train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_rf_pred = rf_model.predict(X_val)

# Evaluate
print("Random Forest Accuracy:", accuracy_score(y_val, y_rf_pred))
print("\nClassification Report:\n", classification_report(y_val, y_rf_pred))
```
