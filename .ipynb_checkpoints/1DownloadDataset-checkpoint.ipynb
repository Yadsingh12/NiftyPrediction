```python
import yfinance as yf

# Define the ticker for Nifty 50
ticker = "^NSEI"

# Download data for the last 10 years
data = yf.download(ticker, start="2013-01-01", end="2023-12-31")

# Keep only the required columns
nifty_data = data[['Open', 'High', 'Low', 'Close', 'Volume']]

# Save to a CSV for future use
nifty_data.to_csv("nifty_10_years.csv")
```
